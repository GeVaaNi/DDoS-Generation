### CIC Flow Processing

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# specify location of directory containing the files
directory = 'containercap-results-CIC'
```

```python
# choose file to be analysed
filename = 'UDP-flood-slow.pcap_Flow.csv'

# define collection of metadata
metadata = ['Flow ID', 'Src IP', 'Src Port', 'Dst IP', 'Dst Port', 'Timestamp', 'Protocol']

# create dataframe
cic_dataframe = pd.read_csv(directory + '/' + filename)
full_df = cic_dataframe.copy()                            # shallow copy
cic_dataframe.drop(metadata, inplace=True, axis=1)        # drop metadata columns
display(cic_dataframe)
```

```python
# print all remaining columns
print(cic_dataframe.columns)
```

```python
# Variable columns
# cic_dataframe = cic_dataframe[cic_dataframe.columns.unique > 1]
dropped_columns  = []
for column in cic_dataframe.columns:
    if len(cic_dataframe[column].unique()) == 1:
        cic_dataframe.drop(column, inplace=True, axis=1) 
        dropped_columns.append(column)
display(cic_dataframe)
```

```python
# print variable columns
print(cic_dataframe.columns)
```

```python
# print fixed columns
print(dropped_columns)
```

```python
# print unique values in every column
for column in cic_dataframe.columns:
    print(column)
    print(cic_dataframe[column].unique())
```

```python
for column in dropped_columns:
    print(column)
    print(full_df[column].unique())
    
```
